# Constantes

- $f_1$ = 150 mm
- $f_2$ = -75 mm
- $f_3$ = 75 mm
- $\phi_1$ = $\phi_2$ = $\phi_3$ = 25.4 mm
- position $L_1$ = 0 mm (référence, $IP$)
- position $OP$ = 197 mm
- distance objet - $IP$ = 1000 mm

# Variables

- position $L_2$ entre 0 et 75 mm
- position $L_3$ entre 75 et 100 mm (à paramétrer selon $L_2$)

# Matrice de transfert

On note $L_i$ la distance d'une lentille par rapport à $L_1$. On note $x_i$ la distance entre la lentille $i$ et le prochain élément optique :

- $M_1$ : distance $OP$ - $L_1$ 
- $M_2$ : lentille $L_1$
- $M_3(x_1)$ : distance $L_1$ - $L_2$
- $M_4$ : lentille $L_2$
- $M_5(x_2)$ : distance $L_2$ - $L_3$
- $M_6$ : lentille $L_3$
- $M_7(x_3(x_1,x_2))$ : distance $L_3$

```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
sp.init_printing()
```

```python
def Mt(d):
    '''Translation'''
    M = np.array([[1, d],
                  [0, 1]])
    return M

def Ml(f):
    '''Thin Lens'''
    M = np.array([[1, 0],
                  [-1/f, 1]])
    return M
```

```python
x1, x2 = sp.symbols("x_1 x_2")

x3 = 197 - x1 - x2
```

```python
Mtr = sp.Matrix(Mt(x3)@Ml(75)@Mt(x2)@Ml(-75)@Mt(x1)@Ml(150)@Mt(1000))
sp.simplify(Mtr)
```

```python
Mtr.det()
```

Les coefficients devant les variables sont en raison d'erreurs numériques. Le déterminant est pratiquement 0.

# Paramétrisation de la distance entre lentille

Soit un rayon $r_i$ initial à $y_i = 0$ et à un angle $\alpha_i$ quelconque, partant du plan $OP$. Soit un rayon $r_f$ final à $y_f = 0$ et à un angle $\alpha_f$ quelconque, arrivant au plan $IP$.

$$
\begin{bmatrix}y_f\\\alpha_f\end{bmatrix} = 
M_{tr}\begin{bmatrix}y_i\\\alpha_i\end{bmatrix}
$$

```python
yi, ai = sp.symbols("y_i alpha_i")

ri = sp.Matrix([yi, ai])

rf = Mtr@ri
r_f = rf.subs(yi, 0)
sol = sp.solve(r_f[0], x2)
sol
```

## Vérification des solutions

La solution donne deux racines, voyons laquelle est la bonne.

```python
# si L_1 = 0 mm
sol[0].subs(x1, 0)
```

```python
# si L_1 = 75 mm
sol[0].subs(x1, 75)
```

```python
# si L_1 = 0 mm
sol[1].subs(x1, 0)
```

```python
# si L_1 = 75 mm
sol[1].subs(x1, 75)
```

La première racine est positive, et respecte les cas extrêmes du placement de $L_2 = x_1$. Cette solution paramétrise donc bien les deux lentilles ensemble.

```python
x_2 = sol[0]
x_2
```

```python
M_tr = Mtr.subs(x2, x_2)
M_tr
```

# Grossissement

Si la matrice de transfert est notée :
$$M_{tr} =
\begin{bmatrix}A&B\\ C&D\end{bmatrix}
$$

il est possible de trouver le grossissement par la valeur $A$.

```python
G = M_tr[0,0]

distances_x1 = np.linspace(0, 75, 1000)

G_list = []
for i in distances_x1:
    G_list.append(G.subs(x1, i))
```

```python
plt.plot(distances_x1, G_list)
```

# Résolution

Pixel pitch : 3.6 $\mu$m

La résolution est la taille d'un pixel dans l'espace objet, trouvable facilement avec la courbe de grossissement.

```python
pixel_pitch = 3.6 # en um
res = np.abs(np.divide(pixel_pitch, G_list))
```

```python
plt.plot(distances_x1, res)
```

# Profondeur de champ

Principe identique au labo 1, sauf que le sténopé est un pixel ici.

```python
def prof_champ(L1):
    # Diamètre des lentilles
    phi = 25.4

    # en haut du pixel
    yf1 = 3.6e-3/2 # en mm
    alphaf1 = -sp.atan((phi-pixel_pitch)/(2*x3))
    
    # en bas du pixel
    yf2 = -3.6e-3/2 # en mm
    alphaf2 = -sp.atan((phi+pixel_pitch)/(2*x3))

    alphai1 = sp.symbols("alpha_i1")
    alphai2 = sp.symbols("alpha_i2")
    
    vf1 = sp.Matrix([yf1, alphaf1])
    vf2 = sp.Matrix([yf2, alphaf2])

    z = sp.symbols("z")

    M_prof = sp.Matrix(Mt(x3)@Ml(75)@Mt(x2)@Ml(-75)@Mt(x1)@Ml(150)@Mt(z))

    solve1 = M_prof.LUsolve(vf1)
    z1 = sp.solve(solve1[0], z)
    z_1 = z1[0].subs(x2, x_2)

    solve2 = M_prof.LUsolve(vf2)
    z2 = sp.solve(solve2[0], z)
    z_2 = z1[0].subs(x2, x_2)

    return z_2.subs(x1, 0) - z_1.subs(x1, 0)
```

```python
prof_champ(10)
```

```python
distances_x1_small = np.linspace(0, 75, 20)

liste_prof_champ = []

for i in distances_x1_small:
    liste_prof_champ.append(prof_champ(i))

liste_prof_champ
```

```python
plt.plot(distances_x1_small, liste_prof_champ)
```

# Facteur de zoom

Le facteur de zoom est donné par le ratio entre le grossissement maximal et le grossissement minimal

```python
X_zoom = np.max(G_list)/np.min(G_list)
X_zoom
```
